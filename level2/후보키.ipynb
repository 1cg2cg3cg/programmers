```
# 64.3 첫 풀이 시도
import collections
import itertools
def solution(relation):
    answer = 0
    used_att = set()

    att = {i for i in range(len(relation[0]))}
    
    check = collections.defaultdict(list)

    for i in att :

        check.clear()

        for idx, r in enumerate(relation) :
            
            check[r[i]] = idx

        if len(check) == len(relation) :
            used_att = used_att | {i}
            answer += 1

    #rint("step1", answer, used_att, answer)

    if len(att-used_att) < 2 :
        return answer

    perms = [list(i) for i in list(itertools.combinations(att-used_att, 2))]
    #rint("P", perms)

    for i in perms :

        check.clear()

        for idx, r in enumerate(relation) :
            check[r[i[0]], r[i[1]]] = idx

        if len(check) == len(relation) :
            used_att = used_att | set(i)
            answer += 1
    
    #rint("step2", answer, used_att, answer)

    if len(att-used_att) < 3 :
        return answer

    perms = [list(i) for i in list(itertools.combinations(att-used_att, 3))]
    #rint("P", perms)

    for i in perms :

        check.clear()

        for idx, r in enumerate(relation) :
            check[r[i[0]], r[i[1]], r[i[2]]] = idx

        if len(check) == len(relation) :
            used_att = used_att | set(i)
            answer += 1

    
    #rint("step3", answer, used_att, answer)

    if len(att-used_att) < 4 :
        return answer

    perms = [list(i) for i in list(itertools.combinations(att-used_att, 4))]
    #print("P", perms)

    for i in perms :

        check.clear()

        for idx, r in enumerate(relation) :
            check[r[i[0]], r[i[1]], r[i[2], r[i[3]]]] = idx

        if len(check) == len(relation) :
            used_att = used_att | set(i)
            answer += 1
    
    #print("step4", answer, used_att, answer)

    return answer
```
